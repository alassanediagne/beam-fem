In this notebook, we are going to implement the following examples of beams under loading

![](examples.png)

```python
beam1 = DiscretizedBeam(nodes=[0, 1, 1.5, 3, 6], E=2e9, I=1.872e-4)
beam1.add_support(coordinate=0, type="Hinge")
beam1.add_support(coordinate=3, type="simple")
beam1.add_udl(interval=[0, 6], magnitude=700)
beam1.add_pointload(coordinate=1, load=4000)
beam1.add_boundary_moment(coordinate=6, value=-20)
beam1.plot()
```

```python
beam2 = DiscretizedBeam(nodes=[0, 4, 8, 12], E=2e9, I=1.872e-4)
beam2.add_support(coordinate=0, type="fixed")
beam2.add_support(coordinate=8, type="simple")

def q2(x):
    return 1000 * np.exp(-x)

beam2.add_load(interval=[0, 12], w=q2)
beam2.plot()
```

```python
beam3 = DiscretizedBeam(nodes=[0, 1, 2.5, 4, 6], E=2e9, I=1.872e-4)
beam3.add_support(coordinate=1, type="simple")
beam3.add_support(coordinate=4, type="simple")
beam3.add_pointload(coordinate=2.5, load=6000)
beam3.add_triangle_load(interval=[4, 6], magnitude=1500)
beam3.plot()
```

```python
beam4 = DiscretizedBeam(nodes=np.linspace(0, 8, 41), E=2e9, I=1.872e-4)
beam4.add_support(coordinate=0, type="fixed")
beam4.add_support(coordinate=8, type="fixed")
beam4.add_udl(interval=[0, 2], magnitude=500)

def w2(x):
    return 500 * (np.sin(np.pi / 2 * x) - 1)

beam4.add_load(interval=[2, 6], w=w2)
beam4.add_pointload(coordinate=5, load=1000)
beam4.add_udl(interval=[6, 8], magnitude=500)
beam4.plot()
```
